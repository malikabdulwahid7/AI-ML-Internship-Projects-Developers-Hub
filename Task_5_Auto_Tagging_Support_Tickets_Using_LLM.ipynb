```python
#  Load Zero-Shot LLM Model

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
#  Support Ticket Dataset

tickets = [
    "My internet is not working since morning",
    "I want to reset my password",
    "Payment failed during checkout process",
    "App is crashing frequently when I open it",
    "How can I update my profile information"
]
```

```python
#  Possible Tags

labels = [
    "Network Issue",
    "Account Issue",
    "Billing Issue",
    "App Bug",
    "General Support"
]
```

```python
#  LLM Tagging Function

def tag_ticket(text):
    result = classifier(text, labels)

    return {
        "ticket": text,
        "top_tag": result["labels"][0],
        "top_3_tags": result["labels"][:3],
        "confidence_scores": result["scores"][:3]
    }
```

```python
#  Run Model on Tickets

results = [tag_ticket(t) for t in tickets]
# Show Results in Table

df = pd.DataFrame(results)
df
```
